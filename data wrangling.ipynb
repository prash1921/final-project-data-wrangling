3. Low Level Data Understanding :
 a. Find count of unique values in location column.
 b. Find which continent has maximum frequency using values
counts.
 c. Find maximum & mean value in 'total_cases'.
 d. Find 25%,50% & 75% quartile value in 'total_deaths'.
 e. Find which continent has maximum
'human_development_index'.
 f. Find which continent has minimum 'gdp_per_capit

```python
unique_val = len(pd.unique(users['location'])) 
conti_maxfreq = users['continent'].value_counts().idxmax()
max_totalcases = users['total_cases'].max()
mean_totalcases = users['total_cases'].mean()
quartile_totaldeaths = users['total_cases'].quantile([0.25,0.50,0.75])
conti_maxindex = users.groupby('continent')['human_development_index'].max().idxmax()
conti_mingdp = users.groupby('continent')['gdp_per_capita'].min().idxmin()

print("Number of unique values in location column:", unique_val)
print("-" * 120)
print("Continent which has maximum frequency:", conti_maxfreq)
print("-" * 120)
print("Maximum Value in 'total_cases':", max_totalcases)
print("-" * 120)
print("Mean Value in 'total_cases':", mean_totalcases)
print("-" * 120)
print("25%, 50% and 75% quartile value in 'total_deaths':", quartile_totaldeaths)
print("-" * 120)
print("Continent which has maximum 'human_development_index':", conti_maxindex)
print("-" * 120)
print("Continent which has minimum 'gdp_per_capita':", conti_mingdp)
```
